Part 2: Real Dataset-Titanic

```python
path = '/kaggle/input/datasets/milkesataye/titanic/train.csv'
df = pd.read_csv(path)
```

         Step1: Dataset Exploration

```python
display(Markdown("Dataset Preview (Top 5 Rows)"))
display(df.head().style.set_caption("First 5 entries of the Titanic dataset"))

display(Markdown("Column Information and Missing Values"))
info_df = pd.DataFrame({
    'Non-Null Count': df.count(),
    'Data Type': df.dtypes,
    'Missing Values': df.isnull().sum()
})
display(info_df.style.background_gradient(cmap='Reds', subset=['Missing Values']))
display(Markdown("Numerical Summary Statistics"))
display(df.describe().style.background_gradient(cmap='YlGnBu').format("{:.2f}"))
```

         Step2: Data Cleaning

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
# Embarked -> mode (taking the first value of the mode series)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
# Cabin -> drop
df.drop(columns=['Cabin'], inplace=True)
# Remove duplicates
df.drop_duplicates(inplace=True)
```

     Step3: Data Analysis 

```python
# 1. Survival Analysis by Gender and Class
gender_survival = df.groupby('Sex')[['Survived']].mean()
class_survival = df.groupby('Pclass')[['Survived']].mean()

print("\n--- Survival Probabilities ---")
# Displaying side-by-side or styled
display(gender_survival.style.background_gradient(cmap='Reds').format("{:.2%}"))
display(class_survival.style.background_gradient(cmap='Blues').format("{:.2%}"))

# 2. Age Analysis by Class
print("\n--- Demographic Analysis ---")
age_analysis = df.groupby('Pclass')[['Age']].mean()
display(age_analysis.style.set_caption("Average Age per Class").format("{:.1f} years"))

# 3. Survival rate by age group 
bins = [0, 12, 18, 60, 100]
labels = ['Child', 'Teenager', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
age_group_survival = df.groupby('AgeGroup', observed=False)[['Survived']].mean()

print("\n--- Survival by Age Group ---")
display(age_group_survival.style.bar(color='#FFA07A').format("{:.2%}"))
```

      Step4: Filtering 

```python
print("\n--- Step 4: Filtering ---")

# 1. Female passengers who survived
female_survivors = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]

# 2. Children who survived (Age < 12)
children_survivors = df[(df['Age'] < 12) & (df['Survived'] == 1)]

# 3. Passengers in 1st class with high survival probability
# We filter for 1st Class Females who survived
high_prob_1st_class = df[(df['Pclass'] == 1) & (df['Sex'] == 'female') & (df['Survived'] == 1)]

# Printing the counts
print(f"Female survivors count: {len(female_survivors)}")
print(f"Children survivors count: {len(children_survivors)}")
print(f"High probability 1st Class survivors: {len(high_prob_1st_class)}")

# Display the first few to show the result of the filter
display(high_prob_1st_class.head().style.set_caption("List of High Survival Probability Passengers (1st Class Females)"))
```
